```python
import pandas as pd
data = pd.read_csv('world-data-2023.csv')
data
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Columnas a excluir según el primer conjunto de nombres
columns_to_exclude = ['Abbreviation', 'Calling Code', 'Capital/Major City', 
                      'Currency-Code', 'Largest city', 'Official language', 
                      'Latitude', 'Longitude']

# Obtener los nombres de las columnas del primer conjunto
columns_to_exclude_names = data.columns[data.columns.isin(columns_to_exclude)]

# Excluir las columnas según el segundo conjunto de nombres
data_cleaned = data.drop(columns=columns_to_exclude_names)

# Inspección inicial de los datos
print(data_cleaned.info())
data_cleaned.columns
```

```python
#Elimino caracteres que impiden usar .astype, como $ % o ,
data_cleaned = data_cleaned.apply(lambda x: x.str.replace('%', '') if x.name in data_cleaned.select_dtypes(include=['object']) else x, axis=0)
data_cleaned = data_cleaned.apply(lambda x: x.str.replace(',', '') if x.name in data_cleaned.select_dtypes(include=['object']) else x, axis=0)
data_cleaned = data_cleaned.apply(lambda x: x.str.replace('$', '') if x.name in data_cleaned.select_dtypes(include=['object']) else x, axis=0)

#Convierto todas las columnas (excepto la columna Country) a float
data_cleaned = data_cleaned.apply(lambda x: x.astype(float) if x.name != 'Country' else x, axis=0)
```

```python
data_free = data_cleaned.dropna()
data_free
```

```python
data_free.columns
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(data_free['Armed Forces size'], data_free['Population'], color='blue')
plt.xlabel('Armed Forces size')
plt.ylabel('Population')
plt.title('Armed Forces size vs Population')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(data_free['Population'], data_free['Urban_population'], color='green')
plt.xlabel('Population')
plt.ylabel('Urban_population')
plt.title('Population vs Urban_population')
plt.grid(True)
plt.show()

```

```python
plt.figure(figsize=(10, 6))
plt.scatter(data_free['Birth Rate'], data_free['Life expectancy'], color='red')
plt.xlabel('Birth Rate')
plt.ylabel('Life expectancy')
plt.title('Birth Rate vs Life expectancy')
plt.grid(True)
plt.show()

```

```python
data_cleaned = data_free.dropna(subset=['Armed Forces size', 'Population'])
data_cleaned
```

```python
Q1 = data_cleaned['Armed Forces size'].quantile(0.25)
Q3 = data_cleaned['Armed Forces size'].quantile(0.75)
IQR = Q3 - Q1

data_cleaned = data_cleaned[(data_cleaned['Armed Forces size'] >= Q1 - 1.5 * IQR) & (data_cleaned['Armed Forces size'] <= Q3 + 1.5 * IQR)]

Q1 = data_cleaned['Population'].quantile(0.25)
Q3 = data_cleaned['Population'].quantile(0.75)
IQR = Q3 - Q1

data_cleaned = data_cleaned[(data_cleaned['Population'] >= Q1 - 1.5 * IQR) & (data_cleaned['Population'] <= Q3 + 1.5 * IQR)]
data_cleaned
```

```python
data_cleaned = data_free.dropna(subset=['Population', 'Urban_population'])
data_cleaned
```

```python
Q1 = data_cleaned['Population'].quantile(0.25)
Q3 = data_cleaned['Population'].quantile(0.75)
IQR = Q3 - Q1

data_cleaned = data_cleaned[(data_cleaned['Population'] >= Q1 - 1.5 * IQR) & (data_cleaned['Population'] <= Q3 + 1.5 * IQR)]

Q1 = data_cleaned['Urban_population'].quantile(0.25)
Q3 = data_cleaned['Urban_population'].quantile(0.75)
IQR = Q3 - Q1

data_cleaned = data_cleaned[(data_cleaned['Urban_population'] >= Q1 - 1.5 * IQR) & (data_cleaned['Urban_population'] <= Q3 + 1.5 * IQR)]
data_cleaned
```

7. Escoge otro par de datos, límpialos, prepáralos y analiza su posible relación. Justifica
utilizando las herramientas vistas y gráficos adecuados.

Voy a elegir el par de datos "Tasa de inscripción bruta en educación primaria" (Gross primary education enrollment (%)) y "Tasa de inscripción bruta en educación terciaria" (Gross tertiary education enrollment (%)) para analizar su posible relación. Comencemos limpiando y preparando estos datos antes de analizarlos.

Primero, vamos a verificar si hay valores nulos en estas columnas y eliminar las filas que contengan valores nulos para cualquiera de estas dos variables:

```python
# Verificar y eliminar filas con valores nulos en las columnas seleccionadas
data_cleaned = data_free.dropna(subset=['Gross primary education enrollment (%)', 'Gross tertiary education enrollment (%)'])
```

Luego, podemos crear un gráfico de dispersión para visualizar la relación entre estas dos variables:

```python
plt.figure(figsize=(10, 6))
plt.scatter(data_cleaned['Gross primary education enrollment (%)'], data_cleaned['Gross tertiary education enrollment (%)'], color='green')
plt.xlabel('Tasa de inscripción bruta en educación primaria (%)')
plt.ylabel('Tasa de inscripción bruta en educación terciaria (%)')
plt.title('Relación entre Tasa de inscripción bruta en educación primaria y terciaria')
plt.grid(True)
plt.show()
```

En este gráfico de dispersión, la tasa de inscripción bruta en educación primaria se representa en el eje x y la tasa de inscripción bruta en educación terciaria se representa en el eje y. Observar patrones o tendencias en la dispersión de los puntos puede indicar una posible relación entre estas variables.

Además, podemos calcular el coeficiente de correlación de Pearson para cuantificar la fuerza y la dirección de la relación entre estas dos variables:

```python
correlation_primary_tertiary = data_cleaned['Gross primary education enrollment (%)'].corr(data_cleaned['Gross tertiary education enrollment (%)'])
print("Coeficiente de correlación entre las tasas de inscripción en educación primaria y terciaria:", correlation_primary_tertiary)
```

El coeficiente de correlación nos dará una medida numérica de la relación entre estas dos variables. Un coeficiente cercano a 1 indicaría una correlación positiva fuerte, mientras que un coeficiente cercano a -1 indicaría una correlación negativa fuerte.

Al analizar el gráfico de dispersión y el coeficiente de correlación, podremos determinar si existe una relación significativa entre la tasa de inscripción bruta en educación primaria y terciaria en los países, lo que nos ayudará a justificar nuestras observaciones utilizando las herramientas analíticas y gráficas adecuadas

```python
import seaborn as sns

# Selecciona las variables relevantes para el Pair Plot
selected_variables = ['Gross primary education enrollment (%)', 'Gross tertiary education enrollment (%)', 'Birth Rate', 'Life expectancy', 'GDP']

# Filtra el DataFrame para incluir solo las filas que no contienen valores nulos en las variables seleccionadas
data_selected = data_cleaned[selected_variables].dropna()

# Crea un Pair Plot
sns.pairplot(data_selected)
plt.suptitle('Pair Plot de Variables Seleccionadas', y=1)
plt.show()
```
